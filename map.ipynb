# 인천 초중고 학교 지도 그리기
#### 공공데이터 포털에서 전국학교데이터를 가지고 온 다음에
#### 필요한 데이터만 선택해서 학교위치를 지도위에 그린다.

```python
import pandas as pd
import folium    # 지도 기능을 제공하는 모듈
```

#### csv 파일 불러오기

```python
# 파일을 읽어서 데이터 프레임으로 저장
data = pd.read_csv('전국초중등학교위치표준데이터.csv', encoding='euc-kr')
```

#### 필요한 데이터만 선택

```python
df = data[['학교명','시도교육청명','학교급구분','위도','경도']]
df
# 인천에 있는 학교만 찾기
school = df[df['시도교육청명'] == '인천광역시교육청']
school
```

```python
# 지도 객체 생성
# Map(지도의 가운데 중심, 줌레벨)
map_= folium.Map(location=[37.48481,126.6277], zoom_start=10)
# 마커를 만들고 지도에 추가
# folium.Marker([37.48481,126.6277], tooltip='인천광역시교육청').add_to(map_)
# map_
```

#### 지도 위에 마커 추가하기

#### 지도 위에 학교 위치 표시

```python
# 데이터프레임의 인덱스 확인
school.index
school.columns
# 데이터 프레임 전체 데이터 확인
for i in school.index:
    name = school.loc[i, '학교명']
    typ = school.loc[i, '학교급구분']
    lat = school.loc[i, '위도']
    lon = school.loc[i, '경도']
    print(i, name, typ, lat, lon)
    # 학교별 마커 만들기
    folium.Marker([lat,lon], tooltip=name).add_to(map_)

```

### 지도 위에 학교 위치 표시
- 초중고에 따라 마커의 색을 표시
- 초등학교 : orange
- 중학교 : green
- 고등학교 : blue

```python
for i in school.index:
    name = school.loc[i, '학교명']
    typ = school.loc[i, '학교급구분']
    lat = school.loc[i, '위도']
    lon = school.loc[i, '경도']

    # 조건문으로 마커의 색 결정하기
    # 학교급 구분에 따라 색 결정
    if typ == '초등학교':
        color = 'orange'
    elif typ == '중학교' :
        color = 'green'
    elif typ == '고등학교' :
        color = 'blue'



    
    # 학교별 마커 만들기
    folium.Marker([lat,lon], tooltip=name,
                  icon=folium.Icon(color=color)).add_to(map_)
```

#### 버스정류장

```python
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding='euc-kr')
df = data[['정류장명','도시명','위도','경도']]
inchen = df[(df['도시명'] == '인천광역시') & (df['정류장명'].str.contains('마을회관'))]
seoul = df[(df['도시명'] == '서울특별시') & (df['정류장명'].str.contains('상가'))]

map_= folium.Map(location=[37.48481,126.6277], zoom_start=10)

combined = pd.concat([inchen, seoul])
combined

for i in combined.index:
    name = combined.loc[i, '도시명']
    lat = combined.loc[i, '위도']
    lon = combined.loc[i, '경도']

    if name == '인천광역시':
        color = 'red'
    elif name == '서울특별시':
        color = 'blue'

    folium.Marker([lat,lon], tooltip=name,
                  icon=folium.Icon(color=color)).add_to(map_)

map_
```
