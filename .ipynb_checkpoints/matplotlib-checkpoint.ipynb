# 맷플롯립
데이터를 시각화하는 모듈로, 그래프와 차트를 제공

```python
import numpy as np
# 맷플롯립은 여러 모듈로 구성
# 그래프 제공
import matplotlib.pyplot as plt
# 그래프를 그릴 때 한글 깨지는 현상 해결
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

## 그래프 그리기

### 선그래프
- plot : 선그래프를 그리는 함수

```python
a = np.arange(1,11)
a
# 숫자배열을 그래프에 그리기
# plot(데이터)
plt.plot(a)
# x축 : 데이터 a
# y축 : plt 가 자동 생성함
```

### 막대그래프
- bar : 화면에 막대 그래프를 그리는 함수

```python
a = ['1반', '2반', '3반','4반']
b = [80,90,85,95]
plt.bar(a,b)
```

## 산점도
- scatter : 화면에 산점도를 그리는 함수

```python
x = [30,25,10,50,25]
y = [9,3,10,6,2]
# x축 : 첫번째 배열에 따라 자동으로 범위가 설정
# y축 : 
plt.scatter(x, y)
```

## 히스토그램
- hist : 화면에 히스토그램을 그리는 함수

```python
# randint : 랜덤 정수 생성 (시작,끝, 개수)
scores = np.random.randint(50,101,100)
scores
```

```python
# plt.hist(데이터, 구간)
plt.hist(scores, bins=10)
# 그래프 꾸미기
plt.title('시험 점수 분포')
plt.xlabel('시험 점수')
plt.ylabel('학생의 수')
```

```python
n = np.arange(10, 21)
n
plt.plot(n)
```

```python
a = ['사과', '바나나', '포도', '오렌지']
b = [50,80,40,70]
plt.bar(a,b)
```

```python
x = [1,2,3,4,5]
y = [40,55,65,80,85]
plt.scatter(x,y)
```

```python
d = np.random.randint(0,11,100)
plt.hist(d, bins=5)
plt.title('유튜브 시청 시간 분포')
plt.xlabel('시청시간')
plt.ylabel('사람수')
```

## 여러 그래프 그리기

- figure(피겨) : 흰 도화지  
   도화지의 사이즈를 설정하거나  
   여러 그래프를 한번에 그릴 때 사용  
- axes(엑시스) : 도화지 안에 들어가는 그래프  

```python
# 1부터 10까지의 숫자 생성
a = np.arange(1,11)
# 선 그래프 그리기
plt.plot(a)
# plot 함수를 쓰면 자동으로 figure(도화지)가 생성된다
```

```python
# 흰 도화지 만들기
# 가로 3인치 세로 3인치 만큼 크기 설정하기
plt.figure(figsize=(3,3))
# 도화지 위에 그래프 그리기
plt.plot(np.arange(1,11))
```

```python
# 도화지 위에 그래프 여러개 그리기(같은셀에서 실행해야)
fig = plt.figure(figsize=(10,3))

# 어떤 모양으로 그릴지 생각
# 두개의 그래프를 한줄에 표시해보자
# add_subplot(행, 열, 위치)
# 1행 2열 중 1번째
ax1= fig.add_subplot(1,2,1)
# 1행 2열 중 2번째
ax2= fig.add_subplot(1,2,2)

students = [1,2,3,4,5]
scores = [80,90,70,85,95]

ax1.plot(scores)
ax2.bar(students,scores)
```

## 여러 그래프 그리기2

```python
# 피겨와 엑시스를 한번에 생성하는 함수
# subplots(행크기, 열크기, 캔버스사이즈)
fig, axs = plt.subplots(1,2,figsize=(10,3)) # 튜플 분해
# axs[0] : 1행 2열의 첫번째 그래프
# axs[1] : 1행 2열의 두번째 그래프
students = [1,2,3,4,5]
scores = [80,90,70,85,95]
axs[0].plot(score)
axs[1].bar(students, scores)
```
