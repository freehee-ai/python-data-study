# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(font='Malgun Gothic')
```

## 시본으로 그래프 만드는 방법
### - 시본은 맷플롭립과 다르게 데이터 프레임을 사용한다.

```python
# 그래프에 넣을 시험 점수 데이터
df = pd.DataFrame({'이름':['영희','철수','민수','지연'],
                  '점수':[85,90,80,100],
                  '반':['1반', '1반', '2반', '2반']})
df
```

### 그래프의 종류
- lineplot : 선 그래프
- barplot :  막대 그래프
- scatterplot : 산점도
- histplot : 히스토그램

### 선 그래프

```python
# lineplot(데이터, x축에 사용할 열, y축에 사용할 열)
# seaborn 으로 그래프를 그릴땐, 위치만 넣어줬던 맷플롭과 다르게
# 매개변수로 값을 설정해야 한다.
sns.lineplot(data=df, x='이름', y='점수')
# 그래프 제목
plt.title('학생별 점수')
```

### 막대 그래프


```python
sns.barplot(data=df, x='이름', y='점수')
plt.title('학생별 점수')
```

### 산점도

```python
# 두 변수의 상관관계를 표현할 때 사용
sns.scatterplot(data=df, x='이름', y='점수')
plt.title('학생별 점수')
```

### 히스토그램
- 많은 양의 데이터로 구간을 나누고, 분포를 표현할 때 사용

```python
arr = np.random.randint(50,101,100)
# arr -> dataframe
df = pd.DataFrame({'점수':arr})
# y 없음
sns.histplot(data=df, x='점수', bins=5)
```

```python
df = pd.DataFrame({'과일':['사과', '바나나', '포도', '딸기'],
                    '판매량': [120,80,150,100]})
fig, axs = plt.subplots(1,3, figsize=(10,3))
sns.barplot(data=df,x='과일',y='판매량',ax=axs[0])
axs[0].set_title('막대')
sns.lineplot(data=df,x='과일',y='판매량',ax=axs[1])
axs[1].set_title('선')
sns.scatterplot(data=df,x='과일',y='판매량',ax=axs[2])
axs[2].set_title('선점도')
```

```python
df = pd.DataFrame({'공부시간':[2,4,6,3,5,7,8],
                     '점수':[55,65,75,60,70,85,90]})
fig, axs = plt.subplots(3,1, figsize=(3,10))
sns.scatterplot(data=df, x='공부시간', y='점수', ax=axs[0])
axs[0].set_title('선점도')
sns.barplot(data=df, x='공부시간', y='점수', ax=axs[1])
axs[1].set_title('막대')
sns.lineplot(data=df, x='공부시간', y='점수', ax=axs[2])
axs[2].set_title('선')
```

```python
df1 = pd.DataFrame({'과일':['사과', '바나나', '포도', '딸기'],
                    '판매량': [120,80,150,100]})
df2 = pd.DataFrame({'공부시간':[2,4,6,3,5,7,8],
                     '점수':[55,65,75,60,70,85,90]})

fig, axs = plt.subplots(2,2, figsize=(10,10))

sns.barplot(data=df1,x='과일',y='판매량', ax=axs[0,0], hue='과일')
axs[0,0].set_title('과일-막대')
sns.lineplot(data=df1,x='과일',y='판매량', ax=axs[0,1])
axs[0,1].set_title('과일-선')

sns.barplot(data=df2,x='공부시간',y='점수', ax=axs[1,0])
axs[1,0].set_title('점수-막대')
sns.scatterplot(data=df2,x='공부시간',y='점수', ax=axs[1,1])
axs[1,1].set_title('점수-산점도')

plt.tight_layout()
```

## 여러 그래프 그리기
### 서브플롯 : 도화지 하나에 여러 그래프를 그리는 기능
- figure(피겨) : 도화지(캔버스)
- axes(엑시스) : 그래프가 그려지는 영역 (서브플롯이라고도 부름)

```python
# 도화지 만들기
# subplots(행열의 크기(행,열), 캔버스의 크기(가로,세로-튜플))
# subplots : 피겨와 엑시스가 한번에 생성됨 (구조분해)
# axs : 배열. 행열에 따라 크기 결정

df = pd.DataFrame({'이름':['영희','철수','민수','지연'],
                  '점수':[85,90,80,100],
                  '반':['1반', '1반', '2반', '2반']})

fig, axs = plt.subplots(1,2, figsize=(10,3))

# 첫번째 영역에 그래프 그리기
sns.barplot(data=df, x='이름', y='점수', ax=axs[0])
axs[0].set_title('학생점수(막대)')
# 두번째 영역에 그래프 그리기
sns.lineplot(data=df, x='이름', y='점수', ax=axs[1])
axs[1].set_title('학생점수(선)')
# 글자가 겹치지않게 레이아웃을 자동으로 조정
plt.tight_layout()
```

```python
df = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '기온':[12,14,13,15,17,18,16]})
fig, axs = plt.subplots(1,2, figsize=(10,3))
sns.lineplot(data=df, x='요일', y='기온', ax=axs[0])
axs[0].set_title('요일별 기온(선)')
sns.scatterplot(data=df, x='요일', y='기온', ax=axs[1])
axs[1].set_title('요일별 기온(산점도)')
```

### 막대 색상, 그래프 제목, xy축 라벨

```python
df = pd.DataFrame({'이름':['영희','철수','민수','지연'],
                  '점수':[85,90,80,100],
                  '반':['1반', '1반', '2반', '2반']})
```

```python
sns.barplot(data=df, x='이름', y='점수', palette='pastel')
```

```python
df = pd.DataFrame({'구': ['서구','서구', '연수구','연수구','남동구','남동구'],
                   '인구(만명)': [45,38,35,28,55,50],
                   '동': ['연희동','가좌동','송도동','옥련동','구월동','논현동']})
fig, axs = plt.subplots(1,2, figsize=(10,3))   

sns.barplot(data=df, x='동',y='인구(만명)', ax=axs[0], hue='구', palette='pastel')
axs[0].text('구월동',55, '55만명', color='red', ha='center', va='bottom')
axs[0].set_title('인천 구.동별 인구 현황(막대)')

sns.lineplot(data=df, x='동',y='인구(만명)', ax=axs[1],
             color='blue', linestyle=':',marker='o')
axs[0].set_title('인천 구.동별 인구 현황(선)')



plt.tight_layout()
```

```python
df= pd.DataFrame({'카테고리':['식품','식품','의류','의류','전자제품','전자제품'],
                  '판매량':[120,150,80,95,200,220],
                  '상품명':['라면','과자','셔츠','신발','노트북','스마트폰']})

fig, axs= plt.subplots(1,2, figsize=(10,3))

sns.barplot(data=df, x='상품명', y='판매량', ax=axs[0], hue='카테고리')
axs[0].set_title('상품별 판매량(막대)')

sns.lineplot(data=df, x='상품명', y='판매량', ax=axs[1],
             color='blue', linestyle=':', marker='o')
axs[1].text('스마트폰',220, '220개', color='red', ha='center', va='bottom')

axs[0].set_title('상품별 판매량(선)')

```
